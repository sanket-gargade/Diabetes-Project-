```python
x = data.iloc[:,0:8]
y = data.iloc[:,-1]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmaxscaler = MinMaxScaler()
```

```python
x_tran = minmaxscaler.fit_transform(x)
x_tran
```

```python
x_tran = pd.DataFrame(x_tran)
x_tran.describe()
```

```python
corr_matrix = data.corr()
```

```python
corr_matrix
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix,annot=True,cmap="coolwarm")
```

```python
from sklearn.model_selection import  train_test_split
```

```python
x_tran,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_tran,y_train)
```

```python
train_pred = rf.predict(x_tran)
train_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_train,train_pred)
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],   # Number of trees in the forest
    'max_depth': [None, 10, 20, 30, 40],  # Maximum depth of trees
    'max_leaf_nodes': [None, 10, 20, 30]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
```

```python
grid_search.fit(x_tran, y_train)
```

```python
print("Best parameters:", grid_search.best_params_)
```

```python
best_rf = grid_search.best_estimator_
```

```python
y_pred = best_rf.predict(x_test)
```

```python
print("Test set accuracy:", accuracy_score(y_test, y_pred))
```

```python
confusion_matrix(y_train,train_pred)
```
